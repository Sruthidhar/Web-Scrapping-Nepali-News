```python

metadata_dict = {
        "Title": [],
        "URL": [],
        "Date": [],
        "Author": [],
        "Author URL": [],
        "Content": [],
        "Category": [],
        "Description": [],
        }

categories_dict = {
        'news': 'https://ekantipur.com/news/',
        'business': 'https://ekantipur.com/business/',
        'world': 'https://ekantipur.com/world/',
        'sports': 'https://ekantipur.com/sports/'
        }

show = False

for category, url in categories_dict.items():
  web_page = http.request('GET', url)
  soup = bs(web_page.data, 'html5lib')

  for row in soup.select("article.normal"):

    title = row.find("h2")

    description = row.find("p").text

    title_text = title.text
    title_link = title.a.get("href")

    # print(title_link)

    if title_link.split(":")[0] != "https":

      title_link = url.split(f"/{category}")[0] + title.a.get("href")

    title_text = title.get_text()
    # print(title_link)

    news_link = http.request('GET', title_link)
    news_soup = bs(news_page.data, 'html5lib')

    date = news_soup.find("time").text
    author_url = news_soup.select_one(".author").a.get("href")
    author_name = news_soup.select_one(".author").get_text()
    news_content = ""

    for content in news_soup.select_one(".row").findAll("p"):
      content = str(content).split(">")[1].split("<")[0]

      if len(content) == 0:
        break
      else:
        content += news_content
      content = news_content

      metadata_dict['Title'].append(title_text)
      metadata_dict['URL'].append(title_link)
      metadata_dict['Date'].append(date)
      metadata_dict['Author'].append(author_name)
      metadata_dict['Author URL'].append(author_url)
      metadata_dict['Content'].append(content)
      metadata_dict['Category'].append(category)
      metadata_dict['Description'].append(None)
      if show:

        print(f"""
              Title: {title_text}, URL: {title_link}
              Date: {date}, Author: {author_name},Category : {category},
              Author URL: {author_url},
              Content: {content}
                      """)
      # news_values.append()
ekantipur_df = pd.DataFrame(metadata_dict, columns=list(metadata_dict.keys()))
ekantipur_df
```
